```python
#While loop

#While test_Expression:
#    Body of while

n = 10
sum = 0
i = 1
while i <= n:
    sum = sum + i
    i = i + 1
    if(i==2):
        print(i)
else:
    print("The program is done")
print("The sum is : ",sum)


```

```python
#break

for val in 'string':
    if val == 'i':
        break
    print(val)
print("End")
```

```python
#Continue

for val in 'string':
    if val == 'i':
        continue
    print(val)
print("End")
```
