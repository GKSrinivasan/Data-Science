```python
data.columns=['Country', 'Year', 'Status', 'Life_expectancy', 'Adult_Mortality',
       'infant_deaths', 'Alcohol', 'percentage_expenditure', 'Hepatitis_B',
       'Measles', 'BMI', 'under-five_deaths', 'Polio', 'Total_expenditure',
       'Diphtheria', 'HIV/AIDS', 'GDP', 'Population',
       'thinness_1to19_years', 'thinness_5to9_years',
       'Income_composition_of_resources', 'Schooling']
```

```python
Country=pd.DataFrame(pd.get_dummies(data.Country))
##Country
data=pd.concat([data,Country],axis=1,sort=None)
##data=pd.concat
data
```

```python
data.Life_expectancy.fillna(data.describe()['Life_expectancy']['std'],inplace=True)
data.Adult_Mortality.fillna(data.describe()['Adult_Mortality']['std'],inplace=True)
data.Hepatitis_B.fillna(data.describe()['Hepatitis_B']['std'],inplace=True)
data.BMI.fillna(data.describe()['BMI']['std'],inplace=True)
data.Polio.fillna(data.describe()['Polio']['std'],inplace=True)
data.Total_expenditure.fillna(data.describe()['Total_expenditure']['std'],inplace=True)
data.Diphtheria.fillna(data.describe()['Diphtheria']['std'],inplace=True)
data.GDP.fillna(data.describe()['GDP']['std'],inplace=True)
data.Income_composition_of_resources.fillna(data.describe()['Income_composition_of_resources']['std'],inplace=True)
data.Schooling.fillna(data.describe()['Schooling']['std'],inplace=True)
```

```python
data.thinness_1to19_years.fillna(data.describe()['thinness_1to19_years']['std'],inplace=True)
data.thinness_5to9_years.fillna(data.describe()['thinness_5to9_years']['std'],inplace=True)
```

```python
df=data.copy()
```

```python
df.drop(['Country','Status'],axis=1,inplace=True)
```

```python
df.info()
```

```python
x=df.drop(['Life_expectancy'],axis=1)
```

```python
y=df['Life_expectancy']
```

```python
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
lm=LinearRegression()
```

```python
lm=lm.fit(x_train,y_train)
```

```python
print("Training feature set size:",x_train.shape)
print("Test feature set size:",x_test.shape)
print("Training variable set size:",y_train.shape)
print("Test variable set size:",y_test.shape)
```

```python
x.info()
```

```python
train_pred = lm.predict(x_train)
```

```python
y_train
```

```python
test_pred = lm.predict(x_test)
```

```python
metrics.mean_squared_error(y_train,train_pred)
```

```python
metrics.mean_squared_error(y_test,test_pred)
```

```python
metrics.mean_absolute_error(y_train,train_pred)
```

```python
np.sqrt(metrics.mean_squared_error(y_train,train_pred))
```

```python
round(metrics.r2_score(y_train,train_pred),3)
```

```python
np.mean(np.abs((y_test-test_pred)/y_test))*100
```

```python
np.mean(np.abs((y_train-train_pred)/y_train))*100
```

```python
pt.figure(figsize=(10,7))
pt.title("Actual vs. predicted",fontsize=25)
pt.xlabel("Actual test set",fontsize=18)
pt.ylabel("Predicted", fontsize=18)
pt.scatter(y_test,test_pred)
```

```python
import seaborn as sns
pt.figure(figsize=(10,7))
pt.title("Histogram of residuals",fontsize=25)
pt.xlabel("Residuals",fontsize=18)
pt.ylabel("Kernel density", fontsize=18)
sns.distplot([y_test-test_pred])
```

```python
pt.figure(figsize=(10,7))
pt.title("Residuals vs. predicted\n",fontsize=25)
pt.xlabel("Predicted house prices",fontsize=18)
pt.ylabel("Residuals", fontsize=18)
pt.scatter(x=test_pred,y=y_test-test_pred)
```
