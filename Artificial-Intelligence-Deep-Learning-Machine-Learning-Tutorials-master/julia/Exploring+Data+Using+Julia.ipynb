## Playing with Data

While our exploration of the data, we found a few problems in the data set, which needs to be solved before the data is ready for a good model. This exercise is typically referred as “Data Munging”. Here are the problems, we are already aware of:

- There are missing values for some variables. We should estimate those values wisely depending on a number of missing values and the expected importance of variables.
- While looking at the distributions, we saw that ApplicantIncome and LoanAmount seemed to contain extreme values at either end. Though they might make intuitive sense, but should be treated appropriately.

In addition to these problems with numerical fields, we should also look at the non-numerical fields i.e. Gender, Property_Area, Married, Education and Dependents to see, if they contain any useful information.

```julia
# Lets check on some missing values
showcols(train)
```

```julia
# Replace all the missing values in columns
train[isna.(train[:LoanAmount]),:LoanAmount] = floor(mean(dropna(train[:LoanAmount])))
```

### What this does is the following

- `train[:LoanAmount]` accesses LoanAmount column from dataframe
- `isna()` - boolean, meaning it returns a true or flase based on missing column
- `train[condition, :column]` - returns the rows of the given column that satisfy the condition (We'll check that later)
- `dropna()` - ignores NA values
- `mean()` - takes mean
- `floor()` performs the floor operation on the value

```julia
?dropna()
```

```julia
# We do that for other columns as well...
#replace 0.0 of loan amount with the mean of loan amount 
 train[train[:LoanAmount] .== 0, :LoanAmount] = floor(mean(dropna(train[:LoanAmount])))
 #replace missing gender with mode of gender values
 train[isna.(train[:Gender]), :Gender] = mode(dropna(train[:Gender]))
#replace missing married with mode value  
 train[isna.(train[:Married]), :Married] = mode(dropna(train[:Married])) 
 #replace missing number of dependents with the mode value
 train[isna.(train[:Dependents]),:Dependents]=mode(dropna(train[:Dependents])) 
 #replace missing values of the self_employed column with mode
 train[isna.(train[:Self_Employed]),:Self_Employed]=mode(dropna(train[:Self_Employed])) 
 #replace missing values of loan amount term with mode value
 train[isna.(train[:Loan_Amount_Term]),:Loan_Amount_Term]=mode(dropna(train[:Loan_Amount_Term]))
#replace credit history missing values with mode  
 train[isna.(train[:Credit_History]), :Credit_History] = mode(dropna(train[:Credit_History]))
```

```julia
using ScikitLearn: fit!, predict, @sk_import, fit_transform!

@sk_import preprocessing: LabelEncoder
@sk_import model_selection: cross_val_score 
@sk_import metrics: accuracy_score
@sk_import linear_model: LogisticRegression
@sk_import ensemble: RandomForestClassifier
@sk_import tree: DecisionTreeClassifier
@sk_import preprocessing: LabelEncoder 
```

### Label Encoding 

```julia
labelencoder = LabelEncoder()
categories = [2 3 4 5 6 12 13]

for col in categories
    train[col] = fit_transform!(labelencoder, train[col])
    if col != 13
        test[col] = fit_transform!(labelencoder, test[col])
    end
end
```

## Build Model 

```julia
function classification_model(model, predictors)
    y = convert(Array, train[:13])
    X = convert(Array, train[predictors])
    X2 = convert(Array, test[predictors])                 

    #Fit the model:
    fit!(model, X, y)

    #Make predictions on training set:
    predictions = predict(model, X)

    #Print accuracy
    accuracy = accuracy_score(predictions, y)
    println("\naccuracy: ",accuracy)

    #5 fold cross validation
    cross_score = cross_val_score(model, X, y, cv=5)
    
    #print cross_val_score
    println("cross_validation_score: ", mean(cross_score))
    
    #return predictions
    fit!(model, X, y)
    pred = predict(model, X2)
    return pred
end
```

```julia
predictors = [:ApplicantIncome, :CoapplicantIncome, :LoanAmount, :Credit_History, :Loan_Amount_Term, :Gender, :Dependents]
```

### Logistic Regression 

```julia
lrmodel = LogisticRegression()
lrpred = classification_model(lrmodel, predictors);
```

### Decision Tree Classifier 

```julia
dtmodel = DecisionTreeClassifier()
dtreepred = classification_model(dtmodel, predictors);
```

### Random Forest Classifier 

```julia
rfmodel = RandomForestClassifier(n_estimators=25, min_samples_split=25, max_depth=7, max_features=1, n_jobs=-1)
rfpred = classification_model(rfmodel, predictors);
```

### Create Submission File 

```julia
pred = map(x-> if x==1 "Y" else "N" end, lrpred)
outdf = DataFrame(Loan_ID=test[:Loan_ID], Loan_Status=pred)
writetable("data/sub.csv", outdf)
```
